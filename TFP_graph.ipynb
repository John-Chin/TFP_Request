```python
import numpy as np
import matplotlib.pyplot as plt

ann_subxn_prices = {

    'The Free Press': 100,
    
    ## News Outlets
    # Conservative
    'Fox': 59.88,
    'Federalist': 40,
    'Reuters': 45,
    'Economist': 62.25, # discount - all 4 years

    'Wall Street Journal': 0, #FREE all 4 years

    # Liberal
    'New York Times': 0, #Free for many students
    'Washington Post*': 12, # discount - for one year
    'CNN': 29.99,
    'Atlantic*':  44.99, # discount - for one year
    

    ## Entertainment
    'Spotify': 65.89,
    'Netflix w Ads': 95.88,
}
labels = list(ann_subxn_prices.keys())
values = list(ann_subxn_prices.values())

x_positions = [i + 0.5 if i == 0 else i + 1 for i in range(len(labels))]

colors = ['#fff2cc', '#cc0000', '#e06666', '#f4cccc', '#f4cccc', '#f4cccc', '#9fc5e8', '#6fa8dc', '#3d85c6', '#3d85c6', '#38761d','#38761d' ]

plt.bar(x_positions, values, width=0.8, color=colors)

# Use the same x_positions for labels
# plt.xticks(x_positions, labels, rotation=90)
custom_labels = [r'$\bf{' + label + '}$' if label == 'The Free Press' else label for label in labels]
plt.xticks(x_positions, custom_labels, rotation=90)


plt.axhline(y=100, color='red', linestyle=':', linewidth=2)
plt.xlabel('news/media outlet')
plt.ylabel('annual subscription cost ($ dollars)')
plt.title("News & Media Subscription Costs for College Students")
plt.figtext(0.5, -0.05, '* discounted price is a one year trial', ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```
